```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de la ecuación diferencial
def dP_dt(P):
    return 0.0225 * P - 0.0003 * P**2

# Función para calcular el porcentaje alcanzado
def percentage_reached(P, P_limit):
    return (P / P_limit) * 100

# Parámetros
P_initial = 25  # Población inicial
P_limit = 75  # Población límite
t_end = 120  # 10 años en meses

# Método de Euler con h = 12 meses (1 año)
h_1 = 12
steps_1 = int(t_end / h_1)
P_1 = P_initial
P_values_1 = [P_1]
time_1 = [0]

for _ in range(steps_1):
    P_1 += h_1 * dP_dt(P_1)
    P_values_1.append(P_1)
    time_1.append(time_1[-1] + h_1)

# Método de Euler con h = 6 meses (0.5 años)
h_2 = 6
steps_2 = int(t_end / h_2)
P_2 = P_initial
P_values_2 = [P_2]
time_2 = [0]

for _ in range(steps_2):
    P_2 += h_2 * dP_dt(P_2)
    P_values_2.append(P_2)
    time_2.append(time_2[-1] + h_2)

# Cálculo del porcentaje alcanzado después de 5 y 10 años
t_5_years = 5 * 12  # 5 años en meses
t_10_years = 10 * 12  # 10 años en meses

# Para h = 1 año
P_5_years_1 = P_values_1[int(t_5_years / h_1)]  # Población después de 5 años
P_10_years_1 = P_values_1[-1]  # Población después de 10 años
percent_5_years_1 = percentage_reached(P_5_years_1, P_limit)
percent_10_years_1 = percentage_reached(P_10_years_1, P_limit)

# Para h = 0.5 años
P_5_years_2 = P_values_2[int(t_5_years / h_2)]  # Población después de 5 años
P_10_years_2 = P_values_2[-1]  # Población después de 10 años
percent_5_years_2 = percentage_reached(P_5_years_2, P_limit)
percent_10_years_2 = percentage_reached(P_10_years_2, P_limit)

# Graficar resultados en una sola gráfica
plt.figure(figsize=(8, 6))

plt.plot(time_1, P_values_1, label='h = 1 año')
plt.plot(time_2, P_values_2, label='h = 0.5 años', color='orange')
plt.xlabel('Tiempo (meses)')
plt.ylabel('Población')
plt.title('Comparación de la Estimación de la Población')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Imprimir resultados
print(f"Con h = 1 año:")
print(f"Población después de 5 años: {P_5_years_1:.2f}")
print(f"Población después de 10 años: {P_10_years_1:.2f}")
print(f"Porcentaje de población límite después de 5 años: {percent_5_years_1:.2f}%")
print(f"Porcentaje de población límite después de 10 años: {percent_10_years_1:.2f}%")

print(f"\nCon h = 0.5 años:")
print(f"Población después de 5 años: {P_5_years_2:.2f}")
print(f"Población después de 10 años: {P_10_years_2:.2f}")
print(f"Porcentaje de población límite después de 5 años: {percent_5_years_2:.2f}%")
print(f"Porcentaje de población límite después de 10 años: {percent_10_years_2:.2f}%")
```
